```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data
df = pd.read_csv(r'C:\Users\Funkaphant\Desktop\Certification\Capstone\Python Project\movies.csv')
```

```python
# Review Data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
df = df.dropna() 
```

```python
# Check column datatypes

df.dtypes
```

```python
# Convert column data type

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Create column with correct year

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicate entries

df['name'].drop_duplicates().sort_values(ascending=False)
```

```python
# Determine correlations to revenue
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Revenue')

plt.xlabel('Budget for film')

plt.ylabel('Gross Revenue')

plt.show()
```

```python
df
```

```python
# Plot budget vs gross with seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
```

```python
# Examing correlation

df.corr()
```

```python
corrlation_matrix = df.corr()

sns.heatmap(corrlation_matrix, annot=True)

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Correlation Metric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Examine Company correlation to gross

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
corrlation_matrix = df_numerized.corr()

sns.heatmap(corrlation_matrix, annot=True)

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Correlation Metric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
